```python
from ultralytics import YOLO
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load a pretrained YOLOv8n model
model = YOLO('./weights/best.pt')

# Define remote image or video URL
source = 'https://images.hindustantimes.com/rf/image_size_630x354/HT/p2/2019/03/30/Pictures/hindustan-vehicles-november-changing-anticipating-commercial-vehicles_a225f784-527e-11e9-b593-df9071890b8d.jpg'

# Run inference on the source
pred = model.predict(source)
result = pred[0].plot()
plt.imshow(result)
```
